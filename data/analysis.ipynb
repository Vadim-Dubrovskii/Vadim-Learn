```python
groupped_df = df.groupby('DR_NDrugs').agg({'DR_Kol': sum})
groupped_df
```

Сделал допонитеную обработку, чтобы посмотреть отсортированное по количесту таблицу.

```python
groupped_df = df.groupby('DR_NDrugs').agg({'DR_Kol': sum}).sort_values(by='DR_Kol', ascending=False)
groupped_df
```

Перезаписываем переменную groupped_df только теперь рассчитываем долю каждой позиции (лекарства) для расчета используем количетво, которое получилось в предыдущем вычислении и сумму всего количества, которую рассчитываем самостоятельно. Далее используем функцию sort_values(ascending=False), чтобы отсортировать по убыванию. Обрати внимание, если вместо False использовать True, то сортировка будет осуществлена по возрастанию.

Теперь используем функцияю cumsum() чтобы рассчитать накопительным итогом количество и сбросить индексы.

```python
groupped_df = groupped_df.cumsum().reset_index()
groupped_df
```

Для дальнейшей работы нам понадобится библиотека numpy - загружаем ее.

```python
import numpy as np
```

Фомриуем новый стоблец дата фрейма groupped_df - ['abc'], для этого используем функцию сравнения из numpy -  np.where (), где определяем каждому продукту - продуктовую группу, если значение индеска до 0,8 - (80%), то это группа A, если меньше 0,95 - (95%), то это группа B, остальное группа C.

```python
groupped_df['abc'] = np.where(groupped_df['DR_Kol'] < 0.8, 'A', np.where(groupped_df['DR_Kol'] < 0.95, 'B', 'C'))
groupped_df
```

Вызовем таблицу с элементами, которые попадают в групп 'B'.

```python
groupped_df[groupped_df['abc'] == 'B']
```

## Как провести многомерный ABC-анализ

Теперь проходим похожий путь, только теперь вместе с количеством мы будем использовать розничную цену.
Переменной groupped_df группируем наш дата фрейм по столбцу лекарства, и суммируем колонки количества и розничной цены продажи.

```python
groupped_df = df.groupby('DR_NDrugs').agg({'DR_Kol': sum, 'DR_CRoz': sum})
groupped_df
```

Повторям, то что делали на предыдущем уроке.
Формируем дополнительный столбец в Дата Фрейме groupped_df - 'rel_kol' в котором рассчитываем долю от общей суммы по столбцу DR_Kol.

```python
groupped_df['rel_kol'] = groupped_df['DR_Kol'] / sum(groupped_df['DR_Kol'])
groupped_df
```

Проводим сортировку дата фрейма groupped_df по возрастанию по колонке 'rel_kol' и перезаписываем его.

```python
groupped_df = groupped_df.sort_values('rel_kol', ascending=False)
groupped_df
```

Формируем дополнительный столбец в дата фрейме groupped_df - cumsum_kol, который рассчитывается куммулитивным итогом от rel_sum.

```python
groupped_df['cumsum_kol'] = groupped_df['rel_kol'].cumsum()
groupped_df
```

Формируем дополнительный стоблец в наш дата фрейма groupped_df - abc_kol, в котором записываем по каждому элементу группу в зависимости от значений кумулятивной суммы элемента.

```python
groupped_df['abc_kol'] = np.where(groupped_df['cumsum_kol'] < 0.8, 'A', np.where(groupped_df['cumsum_kol'] < 0.95, 'B', 'C'))
```

Выводим полученный дата фрейм для просмотра.

```python
groupped_df
```

Теперь выполняем тоже самое для столбца с розницной ценой.
1. Формируем дополнительный столбец rel_roz, как отношение розничной цены элемента к общей сумме розничной цены всех элементов.
2. Сортируем по убыванию столбец rel_roz.
3. Формируем дополнительный столбец cumsum, как куммулятивную сумму столбца rel_roz.
4. Формируем дополнительный столбец abc_roz, в котором каждому элементу присваиваем значение из группы ABC, но основании отбора по столбцу cumsum_roz.


```python
groupped_df['rel_roz'] = groupped_df['DR_CRoz'] / sum(groupped_df['DR_CRoz'])
groupped_df = groupped_df.sort_values('rel_roz', ascending=False)
groupped_df['cumsum_roz'] = groupped_df['rel_roz'].cumsum()
groupped_df['abc_roz'] = np.where(groupped_df['cumsum_roz'] < 0.8, 'A', np.where(groupped_df['cumsum_roz'] < 0.95, 'B', 'C'))
```

Вывозим избранные столбцы, при этом сбрасываем индекс и делаем соритировку по убыванию по столбцу DR_Croz

```python
groupped_df[['DR_CRoz', 'DR_Kol', 'abc_kol', 'abc_roz']].reset_index().sort_values('DR_CRoz', ascending=False)
```

## Как автоматизировать ABC-анализ

Формируем новый дата фрейм из исходной таблицы, который примет на себя элементы, по которым будет производится ABC анализ.
Здесь мы выбираем столбцы DR_NDrugs, DR_Kol, DR_CRoz

```python
df1 = df[['DR_NDrugs', 'DR_Kol', 'DR_CRoz']]
df1.columns
```

Вызываем к просмотру новый дата фрейм

```python
df1
```

В качестве примера, который будем испозовать в дальнейшем создаем новую переменную а, которая формируемт из столбоцов нашего ногового дата фрейма список. а затем исключает из этого списка столбец DR_Kol. Итоге получаем список из наших столбцов без удаленного столбца.

```python
a = list(df1.columns)
a.remove('DR_Kol')
a
```

В качестве примера формируем словарь, каждый столбец в списке a, будет объявлен ключем, а значение будет определено как сумма.

```python
{col: sum for col in a}
```

В качестве примера показываем, что при сложении списка с другим списком будет получен новый список

```python
a + [1, 23]
```

В качестве примера записываем что каждому столбцу col в списке a, может быть использован в качестве суффикса для образования нового столбца.
У в списке a были столбцы DR_NDrugs, DR_CRoz. Используя эти названия и f' мы получили название для нового списка rel_DR_NDrugs, rel_DR_CRoz.

```python
for col in a:
    print(f'rel_{col}')
```

Переменной perform_abc задаем значение дата фрема (наш новый дата фрейм с выбранным наимнованием лекарства, количества и розничной ценой (DR_NDrugs, DR_Kol, DR_CRoz)), в качестве индекса используем R_NDrugs. По нему будет произведен расчет.

```python
perform_abc(df1, index='DR_NDrugs')
```

Теперь записываем функцию, которая позволяет автоматически рассчитывать параметры для определения групп ABC по выбранным столбцам.

```python
def perform_abc(df, index):
    """
    На вход подается датафрейм пандас...
    """
    cols = list(df.columns) # формируем переменную cols, как список столбцов нашего дата фрейма.
    cols.remove(index) # перезаписываем нашу переменную cols, удаляя оттуда значение, которое используется в качестве индекса, и по которому расчеты выполнять не будут
    groupped_df = df.groupby(index).agg({col: sum for col in cols}) # формируем новый дата фрейм и производим группировку по индексируемому столбцу, при этом создаем словарь, в котором ключом выступает название столца, а значением выступает сумма
    for col in cols: # теперь формуем цикл для расчета каждого столбца в списке всех столбцов
        groupped_df[f'rel_{col}'] = groupped_df[col] / sum(groupped_df[col]) # формируем новый столбец, в который записываем долю каждого элемента
        groupped_df = groupped_df.sort_values(f'rel_{col}', ascending=False) # перезаписываем дата фрей проведя сортировку по доле каждого элемента
        groupped_df[f'cumsum_{col}'] = groupped_df[f'rel_{col}'].cumsum() # формируем новый столбец, в котором записываем кумулятивную долю элементов
        groupped_df[f'abc_{col}'] = np.where(groupped_df[f'cumsum_{col}'] < 0.8, 'A', np.where(groupped_df[f'cumsum_{col}'] < 0.95, 'B', 'C')) # формируем столбец, в котором записываем долю в зависимости от кумулятивной суммы
    return groupped_df[cols + [f'abc_{col}' for col in cols]] # возвращаем список со столбцами, которые изначально были в дата фрейме и добавлям новые столбцы, которые содержат префикс 'abc'
```

Задаем значени переменной perform_abc. Получаем новый дата фрейм с указанием групп ABC.

```python
perform_abc(df1, index='DR_NDrugs')
```
