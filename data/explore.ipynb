Функцией ниже делаем группировки по нескольким элементам по разным стоблцам по разным расчетам.
В этом примере, мы группируем по дате 'dt' и номеру чека 'nchk'(Обрати внимание задаем Списком), задаем столбцы, по которым нужно посчитать агрегатные функции 'kol', 'roz', 'zak' (Обрати внимаение задаем их Списком) и считаем агрегатные функции (Обрати внимание задаем условия расчета Словарем, для столбца, который нужно посчитать несколько значений - значения расчет задаем Списком - см. 'roz'):
1. Количетво 'kol' - сумма 'sum'
2. Стоимость розничной продажи 'roz' - сумма 'sum' и максимальное значение 'max'
3. Стоимость закупки 'zak' - сумма'sum'
Еще раз - агрегацию задаем Словарем.
В видеоматериале не корректно указан список столбцов по которым нужно посчитать агрегатные функции, необходимо этот список обернуть дважды в квадратные скобки [[..., ..., ...]].
Проводя группировку мы формируем новый индеск по набору 'dt'и 'nchk'

```python
a = df.groupby(['dt', 'nchk'])[['kol', 'roz', 'zak']].agg({
    'kol': sum, 
    'roz': ['sum', 'max'],
    'zak': sum
})
```

Вызываем функцию с расчетом, чтобы посмотреть, что у нас получилось. Обрати внимание на то, как сформирована таблиц с учетом нового индекса по двум столбцам.

Задача, сфомировать отчет в котром мы можем увидеть прибыль по каждому сотруднику в каждый день продаж.
Будем выполнять эту задачу в два этапа.
На первом этапе сгруппируем дни сотрудников, посчитаем сумму по столбцу цена закупки 'zak' и розничная продажа 'roz'.
Используем функцию a = df.groupby(['dt', 'empl'])['zak', 'roz'].agg(sum)

```python
a = df.groupby(['dt', 'empl'])[['zak', 'roz']].agg(sum)
```

```python
a
```

На втором этапе посчитаем доход, как разницу между Розничной продажей и ценой закупки, функцией a['revenue'] = a['roz'] - a['zak'].
Обрати внимание, что к предыдущей таблице мы указываем новый столбец ['revenue'] и указываем, как он должен быть рассчитан ['roz']-['zak']

```python
a['revenue'] = a['roz'] - a['zak']
```

```python
a
```

Но данный расчет не учитывает количество проданных позиций. Чтобы это сделать нужно сделать много дополнительных шагов.
Но для того чтобы это избежать можно использовать функцию apply и lambda. Обрати внимание на написание функции lambda - ничего сложного.

```python
df.groupby(['dt', 'empl'])[['zak', 'roz', 'kol']].apply(lambda x: sum(x['kol']*(x['roz'] - x['zak'])))
```

Данную фунцию вызвали только для того, чтобы вспомнить названия стобцов. Она носит исключительно справочный характер.

## Как использовать индексы значений и фильтрацию

Здесь учимся искать индексы и делать фильтрацию
Данной функцией вызываем столбцы 'zak', 'roz'

```python
df[['zak', 'roz']]
```

```python
df.zak
```

Метод iloc позволяют выбирать строки со второго по четвертый и столбцы с первого по пятый.

```python
df.iloc[2:5, 1:5]
```

Использование оператора loc позволяет указывать наименование столбца вместо его индекса.

```python
df.loc[2, 'zak']
```

Аналогичный метод at только он работает по пересечнию одной строки и одного столбца.

```python
df.at[2, "zak"]
```

Все тоже самое как в операторе iloc

```python
df.iat[2, 5]
```

Теперь научимся применять фильтры.
Это тоже можно сделать два этапа.
Первым этапом используем булевскую переменную выберем через True False

```python
df['kol'] > 5
```

Теперь используя это распоространим на весь дата фрейм. т.е. отберутся значения где количество больше 5.

```python
df[(df['kol'] > 5)]
```

Добавим для оббора еще одно условие, чтобы розничная цена продажи была больше 50.

```python
df[(df['kol'] > 5) & (df['roz'] > 50)]
```

Данная функция осуществляет фильрацию по типу платежа, где тип платежа на равен ни 15, ни 20 - знак "~" говорит об отрицании. 

```python
df[~df.loc[:, 'pay_type'].isin([15, 20])]
```
