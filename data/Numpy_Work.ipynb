## Как создать стандартизированные многомерные массивы

Функция создаем готовые массивы.
1. Функция eye - создает массив, указываем его размер, особенность, все заполнено нулями, кроме центральной диагонали - там только единицы.
2. Функция zeros - создает массив, особенность - указываем его размер в виде кортежа (дополнительные скобки), все данные заполнены нулями.
3. Функция empty - создает массив, особенность - указываем его размер в виде кортежа (дополнительные скобки), все данные чем то ??? Очень осторожное использование.
4. Функция fill - создает массив, особенность - указыаем его размер в виде кортежа(дополнительные скобки), в качестве заполнления указыаем value.
5. Функция ones - создаем массив, особенность все заполнено единицами.
6. Фунция ones_like() - создает массив заполненный единцами, особенность в скобках нужно указать массив, в качестве примера, из его размера будет создан аналогичный массив заполенный единицами.

```python
np.eye(5,5)
```

```python
np.zeros((2,3,3))
```

```python
np.empty((2,2,2))
```

```python
np.full((2,2,2), fill_value=15)
```

```python
np.ones((2,3,4))
```

## Как работать с многомерным массивом

Если стоит задача, получить значения из числового ряда с одинаковыми диапозоном, в numpy это решается через функцию arange. Указываем начало диапозона, конец диапозона и шаг, с которым нужно выполнить ряд.

```python
np.arange(1,10,0.5)
```

Если стоит задача разбить диапозон на определенное количество точек, тогда лучше использовать функцию linspace. Указаываем начало диапозона, конец диапозона и количество точек, формируется массив с разбивкой по точкам.

```python
np.linspace (0,100,51)
```

## Как индексировать массивы и формировать срезы

Учимся получать данные из матрицы. Вызовем массив c из предыдущих примеров. Посмотрим ее размер.


```python
c
```

```python
c.shape
```

Если будем вызывать поочередно ее ряды, получим одномерные массивы.

```python
c[0]
```

```python
c[1]
```

```python
c[2]
```

Но мы можем выбрать из конкретной матрицы конкретную стороку, например выбираем из второй матрицы вторую сроку.

```python
c[1,1]
```

А если мы хотим из нее выбрать конкретный элемент, наример первый - записываем так:

```python
c[1,1,0]
```

Если мы хотим выбрать определенный диапозон со второй по третью матрицу, со второй по последующую строку, до третьего элемента, то записываем:

```python
c[1:,1:,:2]
```

Если нам нужно указать диапозон в ряду, тогда запишем так:
Для наглядности сначала вытащим целый ряд из массива

```python
c[2,1]
```

А теперь вытащим из ряда элементы с первого по четвертый (помним, что счет начинается с 0, последний элемент, который указан - не берется)

```python
c[2,1,::-1]
```

Если мы хотим перевенуть ряд используем ::-1

```python
c[2,1,::-1]
```

Если мы хотим взять из ряда через второй элемент, тогда используем ::2

```python
c[2,1,::2]
```

Можем усложнить и взять элементы из диапозона

```python
c[2,1,1:4:2]
```

Кроме того, если мы хотим изменить эти элементы мы испозуем следующее:

```python
c[2,1,1:4:2]=100
```

Элементы поменяли значение.

```python
c[2,1]
```

## Что такое бродкастинг

Для общего понимания бродкастинга используем пример. Имеем массив из трех чисел. Умножаем его на какое то число, получаем новый массив, где каждый элемент предыщего массива перемножен на указанное число.

```python
a=np.array([1,2,3], np.int32)
print(a)
x=5
b=a*x
print(b)
```

Нам кажется, что каждый элемент массива был умножен на одно и тоже число, на самом деле "по капотом" по числу  x был создан массив [5,5,5], элементы которого были премножены на первый массив т.е. a=[1,2,3]*x=5=>x=[5,5,5]=[5,10,15]

Усложним задачу, возмем два многомерных массив. Создадим их используя функцию ones и вытащим их размер.

```python
x=np.ones((2,1,5,8))
y=np.ones((2,8,5,1))
x.shape, y.shape
```

Теперь создадим новый массив перемножив значения из двух предыдущих массивов и вытащим их размер. Мы видим, что размер в сводной матрице поменялся, из двух матриц были взяты наибольшие значение векторов. Но это работает только если один из векторов равен единице, в противном случае бродкастинг работать не будет

```python
z= x*y
z.shape
```

```python
x=np.ones((2,2,5,8))
y=np.ones((2,8,5,5))
print(x.shape, y.shape)
z= x*y
print(z.shape)
```

Мы видим, что в этом случае бродкастинг не сработал.
